```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = {
    'Regression':['Linear','Decision','Random Forest','SVR','Ridge','Lasso','KNN'],
    'R2-Score': [94.28245,89.59573,91.86937,90.17146,94.28022,94.27929,91.23239],
    'MSE': [12.279,22.345,17.461,21.108,12.284,12.286,18.830],
    'RMS': [3.504,4.727,4.178,4.594,3.504,3.505,4.339]
}
```

```python
data
```

```python
df=pd.DataFrame(data,index=range(1,8))
df
```

```python
df.style.background_gradient(cmap ='viridis')
```

```python
df.plot(linestyle='dashed', 
     marker='o',markerfacecolor='yellow', markersize=6)
plt.xlabel(list(df.iloc[:,0]))
plt.ylabel('---- Score -----> ')
plt.grid(True,color='violet')
plt.show()
```

```python
colors = ['red','green','orange','yellow','blue','pink','purple']
plt.figure(figsize=(15,6))

ax=sns.barplot(x=df['R2-Score'], y=df.Regression, data=df, palette=colors)
ax.bar_label(ax.containers[0], fmt='%.3f')

plt.xlabel('ACCURACY %', fontweight='bold')
plt.ylabel('REGRESSIONS',fontweight='bold' )
plt.title('Regression vs Accuracy %',fontweight='bold')
plt.show()
```
