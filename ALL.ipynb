```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = {
    'Regression':['Linear','Decision','Random Forest','SVR','Ridge','Lasso','KNN'],
    'R2-Score':[94.28245,89.59573,91.86937,90.17146,94.28022,94.27929,90.86920],
    'MSE': [12.279,22.345,17.461,21.108,12.284,12.286,19.699],
    'RMS': [3.504,4.727,4.178,4.594,3.504,3.505,4.428]
}
```

```python
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.plot.bar(x='Regression', y='R2-Score')
plt.xlabel('Regression')
plt.ylabel('R2 - score')
plt.title('Regressions vs R2-score')
plt.show()
```

```python
colors = ['red','green','orange','yellow','blue','pink','purple']
```

```python
ax = df.plot.barh(x='Regression', y='R2-Score')
plt.xlabel('R2-Score')
plt.ylabel('Regression')
plt.title('graph')


    
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.barplot(x='R2-Score', y='Regression', data=df, palette=colors)

plt.xlabel('ACCURACY %', fontweight='bold')
plt.ylabel('REGRESSIONS',fontweight='bold' )
plt.title('Regression vs Accuracy %',fontweight='bold')
#plt.grid(True,color='b')
plt.show()
```
